# The exchange rate between the euro and the American dollar

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
import matplotlib.ticker as mticker

%matplotlib inline
```

```python
exchange_rates = pd.read_csv("euro-daily-hist_1999_2020.csv")
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

## Data Cleaning

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)
```

```python
exchange_rates.head(3)
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']]
euro_to_dollar.head(3)
```

```python
euro_to_dollar['US_dollar'].value_counts()
```

```python
drop = euro_to_dollar[euro_to_dollar['US_dollar'] == '-']
euro_to_dollar = euro_to_dollar.drop(drop.index,axis=0)
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

## Visualize the evolution of the euro-dollar exchange rate

```python
plt.style.use('seaborn')
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['US_dollar'])
plt.show()
```

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(10).mean()
euro_to_dollar.head(3)
```

```python
plt.style.use('seaborn')
plt.plot(euro_to_dollar['Time'],
         euro_to_dollar['rolling_mean'])
plt.show()
```

```python
euro_to_dollar.info()
```

### We show comparatively how the euro-dollar rate changed under the last three US presidents (George W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)). We can use a line plot.

```python
bush_obama_trump = euro_to_dollar.copy(
                   )[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]
bush = bush_obama_trump.copy(
       )[bush_obama_trump['Time'].dt.year < 2009]
obama = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2009) & (bush_obama_trump['Time'].dt.year < 2017)]
trump = bush_obama_trump.copy(
       )[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]
```

```python
plt.style.use('fivethirtyeight')

plt.figure(figsize=(12,4))
ax1 = plt.subplot(1,3,1)
ax2 = plt.subplot(1,3,2)
ax3 = plt.subplot(1,3,3)

axes=[ax1, ax2, ax3]


### Changes
for ax in axes:
    ax.set_ylim(0.8, 1.6)
    ax.set_yticks([1.0, 1.2, 1.4, 1.6])
    ax.set_yticklabels(['1.0', '1.2','1.4', '1.6'], alpha=0.3)

### Bush - plot

ax1.plot(bush['Time'], bush['rolling_mean'],
        color='#BF5FFF')
ax1.set_xticklabels(['', '2001', '', '2003', '', '2005', '',
                     '2007', '', '2009'], alpha=0.3)
ax1.text(12200, 1.7, 'BUSH', fontsize=18, weight='bold',
        color='#BF5FFF')
ax1.text(12000, 1.65, '(2001-2009)', weight='bold',
        alpha=0.3)

mean = bush['US_dollar'].mean()
ax1.hlines(y=mean, xmin=11400, xmax=14200, color='indigo', linewidth=2)

ax1.text(13300, 1.15, s='mean: %s' % round(mean, 2), weight='bold',
         fontsize=13, c='indigo')
    
### Obama - plot

ax2.plot(obama['Time'], obama['rolling_mean'],
        color='orange')
ax2.set_xticklabels(['', '2009', '', '2011', '', '2013', '',
                     '2015', '', '2017'], alpha=0.3)
ax2.text(15200, 1.7, 'OBAMA', fontsize=18, weight='bold',
        color='orange')
ax2.text(15150, 1.65, '(2009-2017)', weight='bold',
        alpha=0.3)

mean = obama['US_dollar'].mean()
ax2.hlines(y=mean, xmin=14300, xmax=17300, color='darkorange', linewidth=2)

ax2.text(16500, 1.3, s='mean: %s' % round(mean, 2), weight='bold',
         fontsize=13, c='darkorange')
    
### Trump - plot

ax3.plot(trump['Time'], trump['rolling_mean'],
        color='blue')
ax3.set_xticklabels(['2017', '', '2018', '', '2019', '',
                     '2020', '', '2021'], alpha=0.3)
ax3.text(17600, 1.7, 'TRUMP', fontsize=18, weight='bold',
        color='blue')
ax3.text(17560, 1.65, '(2017-2021)', weight='bold',
        alpha=0.3)

mean = trump['US_dollar'].mean()
ax3.hlines(y=mean, xmin=17200, xmax=18700, color='navy', linewidth=2)

ax3.text(17900, 1.18, s='mean: %s' % round(mean, 2), weight='bold',
         fontsize=13, c='navy')

### Title
ax1.text(10800, 1.98, 'EURO-USD rate averaged 1.22 under the last three US presidents',
         fontsize=20, weight='bold')
ax1.text(10800, 1.85, '''EURO-USD exchange rates under George W. Bush (2001 - 2009), Barack Obama (2009-2017),
and Donald Trump (2017-2021)''',
        fontsize=16)
    
plt.show()
```
